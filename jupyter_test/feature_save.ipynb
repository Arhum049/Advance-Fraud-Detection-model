```python
import pandas as pd
```

```python
df_raw = pd.read_csv("fraudTest.csv")
```

```python
city_freq = (
    df_raw['city']
    .value_counts(normalize=True)
    .to_dict()
)

job_freq = (
    df_raw['job']
    .value_counts(normalize=True)
    .to_dict()
)
```

```python
import joblib

joblib.dump(city_freq, "city_freq.pkl")
joblib.dump(job_freq, "job_freq.pkl")

print("Mappings saved successfully")
```
